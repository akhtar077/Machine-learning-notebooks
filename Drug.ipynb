```python
import pandas as pd
```

```python
df=pd.read_csv('drug200.csv')
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_Sex = LabelEncoder()
df['Sex'] = le_Sex.fit_transform(df['Sex'])
le_BP = LabelEncoder()
df['BP'] = le_BP.fit_transform(df['BP'])
le_Cholesterol = LabelEncoder()
df['Cholesterol'] = le_Cholesterol.fit_transform(df['Cholesterol'])
le_Drug = LabelEncoder()
df['Drug'] = le_Drug.fit_transform(df['Drug'])

df.head(6)
```

```python
input = df.drop('Drug',axis='columns')
target = df['Drug']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(input)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_scaled,target,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
lr = LogisticRegression(max_iter=10000)
lr.fit(x_train, y_train)
lr.score(x_test, y_test)
```

```python
svm=SVC(max_iter=10000)
svm.fit(x_train, y_train)
svm.score(x_test, y_test)
```

```python
dt=DecisionTreeClassifier()
dt.fit(x_train, y_train)
dt.score(x_test, y_test)
```

```python
rf=RandomForestClassifier()
rf.fit(x_train, y_train)
rf.score(x_test, y_test)
```

```python
nb=GaussianNB()
nb.fit(x_train, y_train)
nb.score(x_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
lr_scores=cross_val_score(LogisticRegression(solver='liblinear', max_iter=1000), input, target,cv=3)
lr_scores
```

```python
dt_scores = cross_val_score(DecisionTreeClassifier(), input, target, cv=3)
dt_scores
```

```python
svm_scores = cross_val_score(SVC(), input, target, cv=3)
svm_scores
```

```python
rf_scores = cross_val_score(RandomForestClassifier(), input, target, cv=3)
rf_scores
```

```python
nb_scores = cross_val_score(GaussianNB(), input, target, cv=3)
nb_scores
```

```python
print("Logistic Regression Scores:", cross_val_score(LogisticRegression(solver='liblinear', max_iter=1000), input, target,cv=3))
print("Decision Tree Scores:", dt_scores)
print("SVM Scores:", svm_scores)
print("Random Forest Scores:", rf_scores)
print("Naive Bayes Scores:", nb_scores)
```

using Pca to reduce dimensions

```python
from sklearn.decomposition import PCA

pca = PCA(0.95)
X_pca = pca.fit_transform(input)
X_pca.shape
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, target, test_size=0.2, random_state=30)
```

```python
lr=LogisticRegression(max_iter=1000)
lr.fit(X_train_pca, y_train)
lr.score(X_test_pca, y_test)
```

```python
rf=RandomForestClassifier()
rf.fit(X_train_pca, y_train)
rf.score(X_test_pca, y_test)
```

```python
dt=DecisionTreeClassifier()
dt.fit(X_train_pca, y_train)
dt.score(X_test_pca, y_test)
#
```

```python
svm=SVC()
svm.fit(X_train_pca, y_train)
svm.score(X_test_pca, y_test)
```

```python
nb=GaussianNB()
nb.fit(X_train_pca, y_train)
nb.score(X_test_pca, y_test)
```
