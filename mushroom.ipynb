```python
import pandas as pd
```

```python
df=pd.read_csv('mushrooms.csv')
df
```

```python
from sklearn.preprocessing import LabelEncoder

```

```python
for column in df.columns:
    if df[column].dtype == 'object':
        le = LabelEncoder()
        df[column] = le.fit_transform(df[column])

df.head(6)
```

```python
input = df.drop('class',axis='columns')
target = df['class']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(input,target,test_size=0.3,)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
lr = LogisticRegression(max_iter=1000)
lr.fit(x_train, y_train)
lr.score(x_test, y_test)
```

```python
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
dt.score(x_test, y_test)
```

```python
svm = SVC()
svm.fit(x_train, y_train)
svm.score(x_test, y_test)
```

```python
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
rf.score(x_test, y_test)
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
```

```python
nb.fit(x_train, y_train)
nb.score(x_test, y_test )
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(LogisticRegression(solver='liblinear', max_iter=1000), input, target,cv=3)
```

```python
dt_scores = cross_val_score(DecisionTreeClassifier(), input, target, cv=3)
dt_scores
```

```python
svm_scores = cross_val_score(SVC(), input, target, cv=3)
svm_scores
```

```python
rf_scores = cross_val_score(RandomForestClassifier(), input, target, cv=3)
rf_scores
```

```python
nb_scores = cross_val_score(GaussianNB(), input, target, cv=3)
nb_scores
```

```python
print("Logistic Regression Scores:", cross_val_score(LogisticRegression(solver='liblinear', max_iter=1000), input, target,cv=3))
print("Decision Tree Scores:", dt_scores)
print("SVM Scores:", svm_scores)
print("Random Forest Scores:", rf_scores)
print("Naive Bayes Scores:", nb_scores)
```
